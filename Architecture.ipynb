```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(14, 10))

# Data Source Layer
data_sources = [
    ("Source des Donnes\n(Fichier CSV)", (0.05, 0.85), "lightblue"),
    ("Google Colab\nPython", (0.35, 0.85), "lightblue"),
    ("Prtraitement\n(Pandas, NumPy)", (0.65, 0.85), "lightblue")
]

for text, pos, color in data_sources:
    rect = patches.FancyBboxPatch(pos, 0.25, 0.1, boxstyle="round,pad=0.1", fc=color, ec="black")
    ax.add_patch(rect)
    plt.text(pos[0] + 0.125, pos[1] + 0.05, text, ha="center", va="center", fontsize=10, weight='bold')

# Data Ingestion and Staging Layer
staging_layer = [
    ("Nettoyage des Donnes\n(Pandas)", (0.05, 0.7), "lightgreen"),
    ("Normalisation des Donnes\n(MinMaxScaler)", (0.35, 0.7), "lightgreen"),
    ("Squence de Donnes\n(Pandas, NumPy)", (0.65, 0.7), "lightgreen")
]

for text, pos, color in staging_layer:
    rect = patches.FancyBboxPatch(pos, 0.25, 0.1, boxstyle="round,pad=0.1", fc=color, ec="black")
    ax.add_patch(rect)
    plt.text(pos[0] + 0.125, pos[1] + 0.05, text, ha="center", va="center", fontsize=10, weight='bold')

# Data Integration and Transformation Layer
transformation_layer = [
    ("Entranement XGBoost\n(XGBoost)", (0.05, 0.55), "lightcoral"),
    ("Entranement LSTM\n(TensorFlow, Keras)", (0.35, 0.55), "lightcoral"),
    ("Comparaison des Performances\n(Matplotlib)", (0.65, 0.55), "lightcoral")
]

for text, pos, color in transformation_layer:
    rect = patches.FancyBboxPatch(pos, 0.25, 0.1, boxstyle="round,pad=0.1", fc=color, ec="black")
    ax.add_patch(rect)
    plt.text(pos[0] + 0.125, pos[1] + 0.05, text, ha="center", va="center", fontsize=10, weight='bold')

# Serving Predictions
serving_predictions = [
    ("Serveur de Prdictions\n(Streamlit)", (0.25, 0.4), "wheat"),
    ("Accs  la Visualisation\n(Matplotlib, Streamlit)", (0.55, 0.4), "violet")
]

for text, pos, color in serving_predictions:
    rect = patches.FancyBboxPatch(pos, 0.25, 0.1, boxstyle="round,pad=0.1", fc=color, ec="black")
    ax.add_patch(rect)
    plt.text(pos[0] + 0.125, pos[1] + 0.05, text, ha="center", va="center", fontsize=10, weight='bold')

# Arrows
arrow_properties = dict(facecolor='black', edgecolor='black', width=0.005, headwidth=0.03)
connections = [
    ((0.175, 0.85), (0.175, 0.8)),
    ((0.475, 0.85), (0.475, 0.8)),
    ((0.775, 0.85), (0.775, 0.8)),
    ((0.175, 0.7), (0.175, 0.65)),
    ((0.475, 0.7), (0.475, 0.65)),
    ((0.775, 0.7), (0.775, 0.65)),
    ((0.175, 0.55), (0.35, 0.5)),
    ((0.475, 0.55), (0.475, 0.5)),
    ((0.775, 0.55), (0.65, 0.5)),
    ((0.375, 0.4), (0.475, 0.4))
]

for start, end in connections:
    ax.annotate('', xy=end, xytext=start, arrowprops=arrow_properties)

# Titre et ajustements
plt.title("Architecture Schmatique du Projet avec Benchmarking", fontsize=14, weight='bold')
plt.axis('off')
plt.show()




```
